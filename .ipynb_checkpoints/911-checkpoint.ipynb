```python
# Import the Pandas library
# Importing the CSV Data for June
import pandas as pd
import numpy as np

# Create a reference the CSV file desired
csv_path = "data/911_Open_Data_Local_Law_119.csv"

# Read the CSV into a Pandas DataFrame
df = pd.read_csv(csv_path)

df.columns
```

```python
lt = df.loc[df["Description"] == "Combined average response time to life threatening medical emergencies by ambulance and fire units"]


```

```python
lt = lt.reset_index(drop=True)
```

```python
lt.head()
```

```python
lt['Avg. Response Times Minutes'] = round(lt['Response Times'] / 60)
```

```python
lt
```

```python
lt = lt.drop(columns=['Agency'])
```

```python
lt = lt.rename(columns={"Month Name":"Month", "Response Times":"Avg. Response Time Seconds"})
```

```python
lt
```

```python
lt['Description'] = "Life threatening medical emergencies EMS & FDNY"
```

```python
lt
```

```python
lt['Month'] =  pd.to_datetime(lt['Month'])
lt.head()
```

```python
lt = lt.sort_values('Month', ascending=True).reset_index(drop=True)
```

```python
lt['Month'] = lt['Month'].dt.strftime('%m/%Y')
```

```python
lt
```

```python
Manhattan_911  = lt.loc[lt["Borough"] == "Manhattan"].reset_index(drop=True)
Manhattan_911 = Manhattan_911.rename(columns={"# of Incidents":"IncidentsNo", "Avg. Response Time Seconds":"AvgResponseTimesSeconds", "Avg. Response Times Minutes":"AvgResponseTimesMinutes"})
```

```python
Manhattan_911
```

```python
Brooklyn_911  = lt.loc[lt["Borough"] == "Brooklyn"].reset_index(drop=True)
Brooklyn_911 = Brooklyn_911.rename(columns={"# of Incidents":"IncidentsNo", "Avg. Response Time Seconds":"AvgResponseTimesSeconds", "Avg. Response Times Minutes":"AvgResponseTimesMinutes"})
```

```python
Brooklyn_911
```

```python
Bronx_911
```

```python
Queens_911  = lt.loc[lt["Borough"] == "Queens"].reset_index(drop=True)
Queens_911 = Queens_911.rename(columns={"# of Incidents":"IncidentsNo", "Avg. Response Time Seconds":"AvgResponseTimesSeconds", "Avg. Response Times Minutes":"AvgResponseTimesMinutes"})
```
